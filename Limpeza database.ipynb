```python
import pandas as pd
```

```python
df = pd.read_csv("Precipitação anual média de longo prazo em volume.csv")
df.head()
```

```python
# Lista dos países a serem filtrados
countries_to_filter = ['BRA']
```

```python
# Filtrar o DataFrame
df_filtered = df[df['REF_AREA'].isin(countries_to_filter)]
```

```python
# Exibir as primeiras linhas do DataFrame filtrado para verificar
df_filtered.head()
```

```python
df_filtered.columns
```

```python
df_final = df_filtered[['REF_AREA', 'REF_AREA_LABEL', 'TIME_PERIOD', 'OBS_VALUE']]
```

```python
df_final.head()
```

```python
# Zerar index
df_final.reset_index(drop=True, inplace=True)
df_final.head()
```

```python
df_final.rename(columns={'REF_AREA': 'codigo pais', 'REF_AREA_LABEL': 'nome pais', 'TIME_PERIOD': 'ano', 'OBS_VALUE': 'volume mm'}, inplace=True)
df_final.head()
```

```python
df_final.to_csv("Precipitação anual.csv", index=False)
```
